# Importing Dataset

```python
# Importing the dataset
dataset = pd.read_csv('student_performance_dataset.csv')
```

```python
# Loking at the dataset
dataset.head(100)
```

```python
dataset.describe(include='all') # ummary overview of dataset
```

```python
# Looking at unique values
dataset.nunique()
```

```python
#display sum of miss value per column
print("\n","Missing Values per column", dataset.isnull().sum())

#Total 
print("\n","Total:", dataset.isnull().sum().sum())

#display total percentage
print("\n", np.round(dataset.isnull().sum().mean(), 4), " % Total percentage of missing values")
```

```python
# 1. Convert categorical 'Stress' to numerical (Yes=1, No=0) like 'Stress_Yes' and 'Stress_No' containing 0s and 1s
dataset['Stress'] = dataset['Stress'].map({'Yes': 1, 'No': 0})

# 2. Iterate through each column to calculate statistics
for column in dataset.columns:
    print(f"--- Statistics for {column} ---")
    
    # Calculate Mean, Median, and Mode
    mean_val = dataset[column].mean()
    median_val = dataset[column].median()
    mode_val = dataset[column].mode()[0]
    
    # Print the results
    print(f"Mean:   {mean_val:.2f}")
    print(f"Median: {median_val:.2f}")
    print(f"Mode:   {mode_val}")
    print("-" * 30)
```
